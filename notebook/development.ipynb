```python
import numpy as np
import pandas as pd
from matplotlib import collections  as mc
from concorde.tsp import TSPSolver
import matplotlib.pyplot as plt
import time
import pylab as pl
```

```python
#stipple the image

image_np = np.load('data/avatar_edited-stipple.npy')
cities = pd.DataFrame(image_np,columns=["X","Y"])
#solve the TSP.
solver = TSPSolver.from_data(cities.X,cities.Y, norm="EUC_2D")
t = time.time()
tour_data = solver.solve(time_bound=60,verbose=True,random_seed=42)
print(time.time()-t)
print(tour_data.found_tour)
```

```python
lines = [[(cities.X[tour_data.tour[i]],cities.Y[tour_data.tour[i]]),
          (cities.X[tour_data.tour[i+1]],cities.Y[tour_data.tour[i+1]])] for i in range(0,len(cities)-1)]
lc = mc.LineCollection(lines, linewidths=1,colors=(0,0,0))
fig, ax = pl.subplots(figsize=(20,20))
ax.set_aspect('equal')
ax.add_collection(lc)
ax.autoscale()
plt.savefig('output.png')
```
